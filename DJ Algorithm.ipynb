```python
def dj_ora(case,n):
  orac_qc = QuantumCircuit(n+1)
#For balanced Oracle

  if case=="balanced":
    for qubit in range(n):
      orac_qc.cx(qubit,n)

#For Oracle is Constant

  if case=="constant":
    out = np.random.randint(2)
    if out==1:
      orac_qc.x(n)
  orac_gate = orac_qc.to_gate()
  orac_gate.name = "Oracle"
  return orac_gate


```

Now We define the Desutsch-Jozsa algorithm


```python
def dj(n,case = 'random'):
  #set up the input register:
  dj_circuit = QuantumCircuit(n+1,n)
  for qubit in range(n):
    dj_circuit.h(qubit)
    #And set up the output qubit:
    dj_circuit.x(n)
    dj_circuit.h(n)
  #Let's append the oracle gates to our circuit:
  if case == 'random':
    random =np.random.randint(2)
    if random ==0:
        case = 'constant'
    else:
      case = 'balanced'
  oracle = dj_ora(case,n)
  dj_circuit.append(oracle,range(n+1))
  #Finally Perform the H-gates again and measure
  for i in range(n):
    dj_circuit.h(i)
    dj_circuit.measure(i,i)
  return dj_circuit
```
